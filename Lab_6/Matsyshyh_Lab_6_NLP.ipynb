# Створення векторного представлення слів за допомогою GloVe

```python
import numpy as np

def load_glove_model(glove_file):
    glove_model = {}
    with open(glove_file, "r", encoding="utf-8") as f:
        for line in f:
            split_line = line.split()
            word = split_line[0]
            embedding = np.array(split_line[1:], dtype=np.float32)
            glove_model[word] = embedding
    return glove_model

glove_file = "glove.6B.50d.txt"  # Необхідно завантажити файл заздалегідь https://drive.google.com/file/d/1mN2Ee44z6CBkCW4QQ95gHHjQTaSk-fxM/view?usp=sharing
glove_model = load_glove_model(glove_file)
print("Вектор для слова 'language':", glove_model.get("language", "Слово не знайдено"))
```

# Візуалізація ембедінгів використовуючи t-SNE

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from transformers import BertTokenizer, BertModel

# Завантаження моделі BERT і токенізатора
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# Вхідний текст (довше речення для більшої кількості токенів)
sentence = "Natural language processing enables machines to understand human language. Deep learning and transformer models improved NLP applications significantly."

# Токенізація тексту
inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)

# Отримання вихідних векторів (ембедінгів слів)
with torch.no_grad():
    outputs = model(**inputs)
word_embeddings = outputs.last_hidden_state.squeeze(0).numpy()  # Витягнення векторів

# Видалення спеціальних токенів [CLS] та [SEP]
tokens = tokenizer.convert_ids_to_tokens(inputs.input_ids[0])
filtered_tokens = []
filtered_embeddings = []
for token, embedding in zip(tokens, word_embeddings):
    if token not in ["[CLS]", "[SEP]"]:
        filtered_tokens.append(token)
        filtered_embeddings.append(embedding)

# Зменшення розмірності векторів за допомогою t-SNE
perplexity_value = min(10, len(filtered_tokens) - 1)  # Уникнення помилки
tsne = TSNE(n_components=2, perplexity=perplexity_value, random_state=42)
reduced_embeddings = tsne.fit_transform(np.array(filtered_embeddings))

# Візуалізація
plt.figure(figsize=(10, 6))
plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1], marker="o")

# Додавання підписів до точок
for i, token in enumerate(filtered_tokens):
    plt.annotate(token, (reduced_embeddings[i, 0], reduced_embeddings[i, 1]))

plt.title("Візуалізація ембедінгів BERT за допомогою t-SNE")
plt.xlabel("TSNE Dimension 1")
plt.ylabel("TSNE Dimension 2")
plt.show()
```

**1. Написати програму, яка розбиває текст на слова вручну.**

```python
def manual_tokenize(text):
    # Розбиваємо текст на слова, враховуючи пробіли
    words = []
    current_word = ""

    for char in text:
        # Якщо символ — пробіл, додаємо поточне слово до списку (якщо воно не порожнє)
        if char.isspace():
            if current_word:
                words.append(current_word)
                current_word = ""
        # Якщо символ не пробіл, додаємо його до поточного слова
        else:
            current_word += char

    # Додаємо останнє слово, якщо воно є
    if current_word:
        words.append(current_word)

    return words

# Тексти різними мовами
texts = {
    "українська": "Штучний інтелект змінює світ щодня.",
    "англійська": "Artificial intelligence changes the world daily.",
    "китайська": "人工智能每天都在改变世界。",
    "японська": "人工知能は毎日世界を変えています。",
    "корейська": "인공지능은 매일 세상을 변화시킵니다.",
    "німецька": "Künstliche Intelligenz verändert die Welt täglich."
}

# Токенізація для кожної мови
for lang, text in texts.items():
    tokens = manual_tokenize(text)
    print(f"Токенізація ({lang}): {tokens}")
```

```python
import string
import jieba  # Для китайської мови
import konlpy  # Для корейської мови

# Встановлення бібліотек для токенізації
!pip install jieba konlpy

def manual_tokenize_clean(text, lang="українська"):
    # Видаляємо розділові знаки
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Токенізація залежно від мови
    if lang == "українська" or lang == "англійська" or lang == "німецька":
        words = []
        current_word = ""
        for char in text:
            if char.isspace():
                if current_word:
                    words.append(current_word)
                    current_word = ""
            else:
                current_word += char
        if current_word:
            words.append(current_word)
        return words
    elif lang == "китайська":
        return list(jieba.cut(text))
    elif lang == "корейська":
        okt = konlpy.tag.Okt()
        return okt.morphs(text)
    else:
        return text.split()

# Тексти різними мовами (без японської)
texts = {
    "українська": "Штучний інтелект змінює світ щодня.",
    "англійська": "Artificial intelligence changes the world daily.",
    "китайська": "人工智能每天都在改变世界。",
    "корейська": "인공지능은 매일 세상을 변화시킵니다.",
    "німецька": "Künstliche Intelligenz verändert die Welt täglich."
}

# Токенізація з вилученням розділових знаків
for lang, text in texts.items():
    tokens = manual_tokenize_clean(text, lang)
    print(f"Токенізація без розділових знаків ({lang}): {tokens}")
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, MT5Tokenizer, MT5ForConditionalGeneration
from rouge_score import rouge_scorer

# Завантаження моделі T5 для англійської мови
t5_tokenizer = T5Tokenizer.from_pretrained("t5-small")
t5_model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Завантаження багатомовної моделі mT5 для інших мов
mt5_tokenizer = MT5Tokenizer.from_pretrained("google/mt5-small")
mt5_model = MT5ForConditionalGeneration.from_pretrained("google/mt5-small")

# Функція для генерації підсумку
def summarize_text(text, tokenizer, model, lang="англійська"):
    if lang == "англійська":
        input_text = "summarize: " + text
    elif lang == "українська":
        input_text = "summarize in Ukrainian: " + text
    elif lang == "китайська":
        input_text = "summarize in Chinese: " + text
    elif lang == "японська":
        input_text = "summarize in Japanese: " + text
    elif lang == "корейська":
        input_text = "summarize in Korean: " + text
    elif lang == "німецька":
        input_text = "summarize in German: " + text
    else:
        input_text = "summarize: " + text

    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)

    summary_ids = model.generate(
        inputs.input_ids,
        max_length=200,
        min_length=50,
        length_penalty=0.8,
        num_beams=4,
        no_repeat_ngram_size=3,
        early_stopping=True
    )
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    summary = summary.replace("<extra_id_0>", "").replace("<extra_id_10>", "").replace("<extra_id_34>", "").replace("<extra_id_37>", "")
    return summary

# Референсні підсумки (еталони)
reference_summaries = {
    "українська": "У 2050 році в Україні штучний інтелект «Мрія» створив музей із унікальними експонатами та віртуальною подорожжю у козацькі часи, який відвідали мільйони.",
    "англійська": "In 2045, AI 'Elara' helped farmers in England with planting predictions and composed nature-inspired music, later creating a virtual garden that inspired tech-savvy farmers.",
    "китайська": "在2060年，AI“星辰”通过虚拟博物馆和音乐会复兴了中国村庄的传统文化，吸引了全球观众。",
    "японська": "2070年、京都のAI「さくら」が茶道を復活させ、仮想現実で茶会を開催し、世界に広めた。",
    "корейська": "2080년, 부산의 AI '해연'이 바다 생태계를 보호하며 음악을 작곡하고 VR 해저 여행을 제안했다.",
    "німецька": "Im Jahr 2065 schuf die KI 'Licht' in Berlin eine virtuelle Märchenwelt, die alte deutsche Geschichten präsentierte und eine globale Märchenwoche vorschlug."
}

# Генерація підсумків
generated_summaries = {}
for lang, text in texts.items():
    if lang == "англійська":
        summary = summarize_text(text, t5_tokenizer, t5_model, lang)
    else:
        summary = summarize_text(text, mt5_tokenizer, mt5_model, lang)
    generated_summaries[lang] = summary
    print(f"Підсумок ({lang}): {summary}")

# Оцінка підсумків за допомогою ROUGE
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

for lang in texts.keys():
    generated = generated_summaries[lang]
    reference = reference_summaries[lang]
    scores = scorer.score(reference, generated)
    print(f"\nROUGE оцінка для ({lang}):")
    print(f"ROUGE-1: {scores['rouge1']}")
    print(f"ROUGE-2: {scores['rouge2']}")
    print(f"ROUGE-L: {scores['rougeL']}")
```

**• Використати модель gpt-2 для генерації нового контенту на основі

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Завантаження моделі GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2", pad_token_id=tokenizer.eos_token_id)

# Функція для генерації тексту
def generate_text(prompt, max_new_tokens=100):
    inputs = tokenizer(prompt, return_tensors="pt")
    attention_mask = inputs["attention_mask"]  # Додаємо attention_mask
    outputs = model.generate(
        inputs.input_ids,
        attention_mask=attention_mask,
        max_new_tokens=max_new_tokens,  # Кількість нових токенів для генерації
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True
    )
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# Тексти різними мовами (без японської)
texts = {
    "українська": (
        "У 2050 році в Україні з’явився перший у світі музей, створений повністю штучним інтелектом. Цей інтелект, якого назвали «Мрія», не лише розробив архітектуру будівлі, але й створив унікальні експонати: картини, скульптури та навіть вірші."
    ),
    "англійська": (
        "In the year 2045, a small town in England witnessed the rise of an AI named 'Elara,' designed to help farmers. Elara analyzed weather patterns, soil conditions, and historical data to predict the best times for planting and harvesting."
    ),
    "китайська": (
        "在2060年，中国的一个小村庄迎来了一个名为“星辰”的人工智能。星辰被设计用来保护村庄的传统文化。"
    ),
    "корейська": (
        "2080년, 한국의 부산에서 '해연'이라는 인공지능이 개발되었습니다. 해연은 바다 생태계를 보호하기 위해 설계되었으며, 해양 데이터를 분석하여 오염을 예측하고 어부들에게 최적의 어업 시기를 알려주었습니다."
    ),
    "німецька": (
        "Im Jahr 2065 wurde in Berlin eine KI namens 'Licht' entwickelt, die sich darauf spezialisierte, alte deutsche Märchen zu neuem Leben zu erwecken."
    )
}

# Генерація продовження для кожного тексту
for lang, text in texts.items():
    generated = generate_text(text, max_new_tokens=100)
    print(f"\nПродовження ({lang}):")
    print(generated)
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from transformers import BertTokenizer, BertModel

# Завантаження моделі BERT і токенізатора
model_name = "bert-base-uncased"
bert_tokenizer = BertTokenizer.from_pretrained(model_name)
bert_model = BertModel.from_pretrained(model_name)

# Оригінальний і згенерований текст (англійська)
original_text = texts["англійська"]
generated_text = generate_text(original_text, max_new_tokens=100)

# Токенізація обох текстів
original_inputs = bert_tokenizer(original_text, return_tensors="pt", padding=True, truncation=True)
generated_inputs = bert_tokenizer(generated_text, return_tensors="pt", padding=True, truncation=True)

# Отримання ембедінгів
with torch.no_grad():
    original_outputs = bert_model(**original_inputs)
    generated_outputs = bert_model(**generated_inputs)

original_embeddings = original_outputs.last_hidden_state.squeeze(0).numpy()
generated_embeddings = generated_outputs.last_hidden_state.squeeze(0).numpy()

# Токени для обох текстів
original_tokens = bert_tokenizer.convert_ids_to_tokens(original_inputs.input_ids[0])
generated_tokens = bert_tokenizer.convert_ids_to_tokens(generated_inputs.input_ids[0])

# Фільтрація спеціальних токенів ([CLS], [SEP])
filtered_original = [(token, emb) for token, emb in zip(original_tokens, original_embeddings) if token not in ["[CLS]", "[SEP]"]]
filtered_generated = [(token, emb) for token, emb in zip(generated_tokens, generated_embeddings) if token not in ["[CLS]", "[SEP]"]]

# Об'єднання ембедінгів і міток
all_embeddings = np.vstack([emb for _, emb in filtered_original] + [emb for _, emb in filtered_generated])
all_tokens = [token for token, _ in filtered_original] + [token for token, _ in filtered_generated]
labels = ["original"] * len(filtered_original) + ["generated"] * len(filtered_generated)

# Зменшення розмірності за допомогою t-SNE
perplexity_value = min(10, len(all_embeddings) - 1)
tsne = TSNE(n_components=2, perplexity=perplexity_value, random_state=42)
reduced_embeddings = tsne.fit_transform(all_embeddings)

# Візуалізація
plt.figure(figsize=(12, 8))
colors = {"original": "blue", "generated": "red"}
for i, (x, y) in enumerate(reduced_embeddings):
    plt.scatter(x, y, c=colors[labels[i]], label=labels[i] if i == 0 or labels[i] != labels[i-1] else "")
    plt.annotate(all_tokens[i], (x, y), fontsize=8)

plt.title("Візуалізація ембедінгів оригінального і згенерованого тексту (англійська)")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.legend()
plt.show()
```

```python
import gensim.downloader as api
import numpy as np
from scipy.spatial.distance import cosine
import matplotlib.pyplot as plt

# Завантаження переднавченої моделі FastText для англійської мови
print("Завантаження моделі FastText...")
fasttext_model = api.load("fasttext-wiki-news-subwords-300")  # Модель на 300 вимірів

# Список слів (наприклад, тематика тварин, але одне слово не пов’язане)
word_list = ["dog", "cat", "bird", "fish", "car"]

# Отримання векторів слів
word_vectors = {word: fasttext_model[word] for word in word_list}

# Обчислення середньої схожості для кожного слова
similarities = {}
for word in word_list:
    word_vec = word_vectors[word]
    # Обчислюємо косинусну схожість із усіма іншими словами
    avg_similarity = np.mean([
        1 - cosine(word_vec, word_vectors[other_word])
        for other_word in word_list if other_word != word
    ])
    similarities[word] = avg_similarity

# Визначення найменш схожого слова
least_similar_word = min(similarities, key=similarities.get)
print(f"Найменш схоже слово: {least_similar_word} (середня схожість: {similarities[least_similar_word]:.3f})")

# Візуалізація схожості
plt.bar(similarities.keys(), similarities.values(), color="orange")
plt.title("Середня косинусна схожість слів у списку (FastText)")
plt.ylabel("Середня схожість")
plt.xlabel("Слова")
plt.show()
```
